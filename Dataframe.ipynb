**Data Frame objects and Example**

```python
# Importing the pandas library and gives it an alias pd
import pandas as pd
```

```python
#Creating an object of the dataframe

student_register = pd.DataFrame()
```

```python
# Assigning values to rows and columns to the dataframe
student_register["Name"] = ["John","Joe","Mary","Tim"]
student_register["Age"] = [15,18,16,19]
student_register["Student"] = [True,False,True,False]
```

```python
student_register
```

```python
#Adding a new row to the Dataframe
new_person = pd.Series(["Jack",17,True],index = ["Name","Age","Student"])
```

```python
student_register._append(new_person,ignore_index= True)
```

```python
# Showing Shape , info and corr function

print ("Shape of dataframe: ")
print (student_register.shape)
print ("Info dataframe: ")
print (student_register.info())
print ("Correlation of dataframe: ")
print (student_register[["Age","Student"]].corr())
print (student_register.corr(numeric_only=True))
```

```python
# All statistical operations
#Describe
print ("Describe dataframe: ")
print (student_register.describe())
```

```python
# Dropping a column
# Example , lets drop Age from the column
student_register = student_register.drop('Age', axis=1) #axis = 1 is for column
print(student_register)
```
