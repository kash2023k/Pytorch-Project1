**Generator & Decoder**

```python
def simplegeneratorFun():
  yield 1
  yield 2
  yield 3
```

```python
for value in simplegeneratorFun():
  print(value)
```

```python
# A simple generator for fibonacci numbers

def fib(limit):
  a , b = 0 , 1
  while a < limit:
    yield a
    a,b = b, a+b
```

```python
x= fib(10)
```

```python
for a in x:
  print(a)
```
