**Examples of Iterators in Python**

```python
my_tuple = ["Apple","Banana","Guava"]
my_iterator = iter(my_tuple)

print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
```

```python
# Iterating using the for loop

for x in my_tuple:
  print(x)
```

```python
# Iterating through a string
fruit = "Banana"
for x in fruit:
  print(x)
```
