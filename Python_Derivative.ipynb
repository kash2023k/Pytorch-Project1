```python
import numpy as np
import torch
```

```python
x = torch.tensor(2.0,requires_grad=True)
y=x**2+3*x+2
z= y.backward()
derivate = x.grad
print("Derivative :",derivate.item())
```

    **Computational Graph of the code above**
             ┌────────────┐
             │   x = 2.0   │
             │ (requires_grad)
             └───────┬────┘
                     │
          ┌──────────┼──────────┐
          │          │          │
      (square)     (multiply)   │
       x ** 2         3 * x      │
          │          │          │
          └──────┬───┘          │
                 │              │
           x2 = x^2         term2 = 3x
                 │              │
                 └──────┬───────┘
                        │
              add constant (+2)
                        │
                        ▼
               ┌────────────┐
               │    y        │
               └───────┬────┘
                       │
               y.backward()
                       │
                       ▼
         PyTorch computes dy/dx and fills x.grad
