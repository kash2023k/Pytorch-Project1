**TENSORS**

* A torch. Tensor is a multi-dimensional matrix containing elements of a single data type.

* Similar To Numpy Arrays, but full of fun things that make them work better on GPU's (vs regular CPU's).

* Default data type of float32.

* More suitable for deep learning than a numpy array.

```python
import torch
import numpy as np
```

**TENSOR OPERATIONS**

```python
mytorch = torch.arange(10)
mytorch
```

```python
# Reshape the data - Reshape and View
mytorch = mytorch.reshape(2,5)
mytorch
```

```python
# Reshape if you dont know the number of items with -1
mytorch2 = torch.arange(10)
mytorch2
```

```python
# Using reshape
mytorch2 = mytorch2.reshape(2,-1)
mytorch2
```

```python
# Using View
mytorch3 = torch.arange(10)
mytorch3
```

```python
mytorch4 = mytorch3.view(2,5)
mytorch4
```

```python
# The tensor changes when reshaped with reshape and view

mytorch3[2] = 100
mytorch3
mytorch4
```

```python
# Slices allows you to grab a section of the tensor
mytorch5 = torch.arange(10)
mytorch5
```

```python
# Grab a specific item
mytorch5[7]
```

```python
# Grab a slice
mytorch5[2:7]
```

```python
mytorch6 = mytorch5.reshape(5,2)
mytorch6
```

```python
# Grabbing a second column
mytorch6[:,1] # : denotes from start and 1 denotes the column 2
```

```python
# Returning a column keeping the shape
mytorch6[:,1:]
```

**TENSOR MATH OPERATIONS**

*   Add , Subtract , Multiply , Divide , Remainders , Exponents
*   Short hand and Long Hand
* Reassignment



```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# Addition Short Hand

tensor_a + tensor_b
```

```python
# Addition Long Hand

torch.add(tensor_a,tensor_b)
```

```python
# Subtract Short hand

tensor_a - tensor_b
```

```python
# Subtract Long Hand

torch.sub(tensor_a,tensor_b)
```

```python
# Multiply Short hand

tensor_a * tensor_b
```

```python
# Multiply Long Hand

torch.mul(tensor_a,tensor_b)
```

```python
# Divide Short Hand

tensor_a / tensor_b
```

```python
# Divide Long Hand

torch.div(tensor_a,tensor_b)
```

```python
# Remainder Short Hand

tensor_b % tensor_a
```

```python
# Remainder Long Hand

torch.remainder(tensor_b,tensor_a)
```

```python
# Exponents / Power short hand
tensor_a ** tensor_b
```

```python
# Exponents / Power long hand

torch.pow(tensor_a,tensor_b)
```

```python
# Another way
tensor_a.add(tensor_b)
```

```python
# Reassignment_ | Want to assign value of tensor_a + tensor_b to tensor_a

# Long hand

# tensor_a = tensor_a + tensor_b

#Short Hand with an _

tensor_a.add_(tensor_b)

```
